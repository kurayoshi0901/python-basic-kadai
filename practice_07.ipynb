```python
#課題後の実践1:
#2. リストに整数・浮動小数点数・文字列をそれぞれ格納してください。その結果をprintしてください。

subjects = ["英語", "数学","理科"]
subjects_average_score = [70, 80, 90]
standard_deviation = [2.4, 4.5, 3.3]

print(subjects)
print(subjects_average_score)
print(standard_deviation)

#3. リストに入れた各値を変数に格納し、それぞれの型(type)をprintしてください。
for i in range(0, 3):
    subject_type = subjects[i]
    print(type(subject_type))

for l in range(0, 3):
    subjects_average_score_type = subjects_average_score[l]
    print(type(subjects_average_score_type))


for m in range(0, 3):
    standard_deviation_type = standard_deviation[m]
    print(type(standard_deviation_type))
```

```python
#実践2
'''数値を入力し、その数値が正の数なら"正の数(2)です"、
負の数なら"負の数(-5)です"、0なら"0です"と出力するプログラムを作成してください。
()内の数値はfで出力してください。'''

import random
numbers = random.randint(-3,3 )

if numbers == 0:
    print(f"{numbers}です")

elif numbers > 0:
    print(f"正の数({numbers})です")

else:
    print(f"負の数{numbers}です")
```

```python
#実践3
'''- 西暦の年を入力し、その年がうるう年かどうかを判定するプログラムを作成してください。うるう年の条件は以下です。
    - 西暦年が4で割り切れる年はうるう年
    - ただし、100で割り切れる年はうるう年ではない
    - しかし、400で割り切れる年はうるう年 となります。
- ヒント: 条件文の優先事項が何かに注意してください。'''

import random
years = random.randint(0,2024 )

if years % 4 != 0:
    print(f"西暦{years}年です。うるう年ではありません")

elif years % 100 == 0 and years % 400 != 0:
    print(f"西暦{years}年です。うるう年ではありません")
else:
    print(f"西暦{years}年です。うるう年です")
```
