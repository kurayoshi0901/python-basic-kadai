```python
#実践1:
''''リストの要素を1つずつ表示するプログラムを作成してください。ただし、要素が'q'の場合は繰り返しを終了してください。'''
# そのほか：文字列のfor

letters = ["a", "c", "q",  "z"]
for i in range(0,50):
    print(letters[i])
    if letters[i] == "q":
        break
```

```python
#実践2:
'''* 文字列(英語)を入力として受け取り、その文字列内の母音の数を数えるプログラムを作成してください。ただし、大文字と小文字は区別しないものとします。'''

words = "abcdefghijklmnopqrstuvwxyz"

numbers_of_a = words.count("a")
numbers_of_i = words.count("i")
numbers_of_u = words.count("u")
numbers_of_e = words.count("e")
numbers_of_o = words.count("o")

numbers_of_vowel = numbers_of_a + numbers_of_i + numbers_of_u + numbers_of_e +numbers_of_o

print(f"母音の数は{numbers_of_vowel}です")
```

```python
#実践3
'''* 2つのリストを作成し、forを重ねて実行してください。その結果をprintしてください。'''

list1 = ["a", "b", "c"]
list2 = ["e", "f", "g"]

for i in list1:
    for m in list2:
        print(f"{i}{m}")
```

```python
#実践4:
'''2つのリストをforで回し、それらをtupleとして
新しいリストへ格納するプログラムを作成してください。'''

list1 = [1, 2, 3, 4]
list2 = ['a', 'b', 'c', 'd']
new_list = []

for l1, l2 in zip(list1, list2):
    new_list.append((l1, l2))

print(new_list)
```

```python
#実践5:
'''* リストに数字を複数入力し、入力した数字の合計値が100を超えたら
繰り返しを終了するプログラムを作成してください。
inputとint変換を使うとわかりやすいです。'''

import random
numbers = random.randint(1,50)
numbers_sum_list = []
numbers_sum_list.append(numbers)
numbers_sum = numbers

while numbers_sum <= 100:
    numbers_sum_list.append(numbers)
    numbers = random.randint(1,50)
    numbers_sum = numbers_sum + numbers

print(numbers_sum_list)
print(f"合計{numbers_sum}")
```

```python
#実践6:
'''* 与えられたディクショナリーから、
値が偶数のキーと値のペアのみを新しいディクショナリーとして
出力するプログラムを作成してください。'''

subject = {80: "英語", 77: "数学", 88: "理科", 63:"国語"}
subject_list = {}

for key, value in subject.items():
    if key % 2 == 0:
        subject_list[key] = value
print(subject_list)
```
